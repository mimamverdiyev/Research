# Paired T-test assumptions

```python
# Importing Shapiro, Anderson library
from scipy.stats import shapiro, anderson

# Import Levene's test
from scipy.stats import levene, f_oneway
```

## **Normality**

```python
# Checking Normal Distribution pre session data

# 'pre_var' is the pre session data
# Shapiro-Wilk test
stat_shapiro, p_value_shapiro = shapiro(pre_var)
print("Shapiro-Wilk Test:")
print("Test Statistic:", stat_shapiro)
print("P-value:", p_value_shapiro)

# Anderson-Darling test
result_anderson = anderson(pre_var)
print("\nAnderson-Darling Test:")
print("Test Statistic:", result_anderson.statistic)
print("Critical Values:", result_anderson.critical_values)
print("Significance Levels:", result_anderson.significance_level)
```

```python
# Checking Normal Distribution post session data

# 'post_var' is the post session data
# Shapiro-Wilk test
stat_shapiro, p_value_shapiro = shapiro(post_var)
print("Shapiro-Wilk Test:")
print("Test Statistic:", stat_shapiro)
print("P-value:", p_value_shapiro)

# Anderson-Darling test
result_anderson = anderson(post_var)
print("\nAnderson-Darling Test:")
print("Test Statistic:", result_anderson.statistic)
print("Critical Values:", result_anderson.critical_values)
print("Significance Levels:", result_anderson.significance_level)
```

```python
# Checking Normal Distribution follow session data

# 'follow_var' is the follow session data
# Shapiro-Wilk test
stat_shapiro, p_value_shapiro = shapiro(follow_var)
print("Shapiro-Wilk Test:")
print("Test Statistic:", stat_shapiro)
print("P-value:", p_value_shapiro)

# Anderson-Darling test
result_anderson = anderson(follow_var)
print("\nAnderson-Darling Test:")
print("Test Statistic:", result_anderson.statistic)
print("Critical Values:", result_anderson.critical_values)
print("Significance Levels:", result_anderson.significance_level)
```

## **Homogeneity of Variances**

```python
# Levene's test for equality of variances pre vs post
statistic_levene, p_value_levene = levene(pre_var, post_var)
print(f"Levene's Test: Statistic = {statistic_levene}, p-value = {p_value_levene}")

# F-test for equality of means
statistic_f, p_value_f = f_oneway(pre_var, post_var)
print(f"F-Test: Statistic = {statistic_f}, p-value = {p_value_f}")
```

```python
# Levene's test for equality of variances post vs follow
statistic_levene, p_value_levene = levene(post_var, follow_var)
print(f"Levene's Test: Statistic = {statistic_levene}, p-value = {p_value_levene}")

# F-test for equality of means
statistic_f, p_value_f = f_oneway(post_var, follow_var)
print(f"F-Test: Statistic = {statistic_f}, p-value = {p_value_f}")
```

```python
# Levene's test for equality of variances pre vs post
statistic_levene, p_value_levene = levene(pre_var, follow_var)
print(f"Levene's Test: Statistic = {statistic_levene}, p-value = {p_value_levene}")

# F-test for equality of means
statistic_f, p_value_f = f_oneway(pre_var, follow_var)
print(f"F-Test: Statistic = {statistic_f}, p-value = {p_value_f}")
```

## **Post**

```python
import pandas as pd
import re


# Function to categorize sentences into 'search', 'boolean', or 'search_boolean'
def categorize_sentence(sentence):
    has_research = re.search(r'\bresearch\b', sentence, re.IGNORECASE)
    has_search = re.search(r'\bsearch\w*\b|\bdatabase\w*\b', sentence, re.IGNORECASE)
    has_boolean = re.search(r'\bboolean\b|\b(?:and|or|not)(?:\s+(?:and|or|not)\b)*' , sentence, re.IGNORECASE)
    has_evaluation = re.search(r'\breliable\b|\baccuracy\b', sentence, re.IGNORECASE)


    if has_research:
      return 'research'
    elif has_search:
      return 'search'
    elif has_boolean:
      return 'search'
    elif has_evaluation:
      return 'research'
    else:
      return 'search'

# Apply the categorize_sentence function to create a new 'category' column
post['category'] = post['top_skills'].apply(categorize_sentence)

# Display the resulting DataFrame
post[["Email", "top_skills", "category"]]
```

```python
post['category'].value_counts()
```
